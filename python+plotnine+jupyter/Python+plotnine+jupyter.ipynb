### Color scheme - Color Blind

```python
# These are the "Tableau Color Blind" colors as Hex.  
 color=['#006BA4','#FF800E','#ABABAB','#A2C8EC','#595959']
```

### Getting the data

```python
cars=pd.read_csv("cars-sample.csv")
```

```python
ggplot(cars, aes('Weight', 'MPG', color='Manufacturer',size='Weight')) + geom_point(alpha=0.5) + scale_color_manual(values=color)


```
