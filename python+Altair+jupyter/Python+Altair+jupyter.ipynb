### Library

```python
import altair as alt
from altair import Chart
import pandas as pd
alt.renderers.enable('notebook')
```

### Getting the data

```python
cars=pd.read_csv("cars-sample.csv")
```

### Altair plotting

```python
chart=Chart(cars)
xrange=(1500,5000)
chart.mark_circle().encode(
    x=alt.X('Weight',scale=alt.Scale(zero=False), axis=alt.Axis(tickCount=4)),
    y=alt.Y('MPG',scale=alt.Scale(zero=False),axis=alt.Axis(tickCount=4)),
    color='Manufacturer',
    size='Weight'
).configure_mark(
    opacity=0.5
)
```
