### Library

```python
import seaborn as sns
import pandas as pd
```

### Getting the data

```python
cars=pd.read_csv("cars-sample.csv")
```

### Seaborn plotting

```python
sns.axes_style("ticks")
sns.set_style("ticks")
plot=sns.relplot(x='Weight', y='MPG',
           data=cars, hue='Manufacturer', 
            size='Weight',sizes=(40,180),alpha=.5,legend="brief")
plot.ax.set_xticks(range(2000,5001,1000))
plot.ax.set_yticks(range(10,41,10))
plot.ax.grid(linestyle='-', linewidth='0.5', color='lightgrey')
plot.ax.spines["bottom"].set_color("lightgrey")    
plot.ax.spines["left"].set_color("lightgrey") 
```
